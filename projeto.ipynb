```python
import pandas as pd
```

```python
lego = pd.read_csv('datasets/lego_sets.csv')
parent = pd.read_csv('datasets/parent_themes.csv')
```

```python
lego = lego[lego['set_num'].notna()]
parent = parent.rename(columns={'name': 'parent_theme'})
```

```python
junto = pd.merge(lego, parent)
```

```python
junto_t = junto[junto['is_licensed'] != False]
```

```python
total = junto_t.count()['name']
```

```python
starwars = junto_t[junto_t['parent_theme'] == 'Star Wars'].count()['name']
```

```python
print(f'{(starwars/total*100).round(2)}% of all sets where Star Wars themed.')
```
